```python
import pandas as pd
import numpy as np
from IPython.display import display
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_percentage_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy import stats
import seaborn as sns

file = 'auto.csv'


pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

# Предобработка данных
def preprocessing(file):
    data = pd.read_csv(file, index_col=0, encoding='cp1251')
    data['Модель'] = data['Марка'] + ' ' + data['Модель']
    data = data.drop(
        ['Марка', 'Класс', 'Привод', 'Тип двигателя', 'Количество мест', 'Расположение двигателя', 'Марка топлива'],
        axis=1)
    data.rename(columns={'Цена': 'Цена, руб', 'Объем двигателя': 'Объем двигателя, л', 'Л/С': 'Мощность, л.с',
                         'Расход': 'Расход топлива, л/100 км', 'Разгон': 'Разгон до 100 км/ч, с',
                         'Макс. скорость': 'Макс. скорость, км/ч', 'Объем багажника': 'Объем багажника, л',
                         'Клиренс': 'Клиренс, мм', 'Объем топливного бака': 'Объем топливного бака, л',
                         'Полная масса': 'Полная масса, кг'}, inplace=True)
    data['Год'] = data['Год'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Цена, руб'] = data['Цена, руб'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Мощность, л.с'] = data['Мощность, л.с'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Макс. скорость, км/ч'] = data['Макс. скорость, км/ч'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Объем багажника, л'] = data['Объем багажника, л'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Объем двигателя, л'] = data['Объем двигателя, л'].str.replace(r'\D+', '', regex=True).astype(float) / 10
    data['Расход топлива, л/100 км'] = data['Расход топлива, л/100 км'].str.replace(r'\D+', '', regex=True).astype(
        float) / 10
    data['Разгон до 100 км/ч, с'] = data['Разгон до 100 км/ч, с'].str.replace(r'\D+', '', regex=True).astype(float) / 10
    data['Продано за 2023'] = data['Продано за 2023'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Клиренс, мм'] = data['Клиренс, мм'].str.replace(r'\D+', '', regex=True).astype(int)
    data['Объем топливного бака, л'] = data['Объем топливного бака, л'].str.replace(r'\D+', '', regex=True).astype(int)
    display(data)
    # Визулизация важных данных датасета
    x=data['Продано за 2023']
    y=data['Модель']
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel('Продано за 2023')
    ax.set_ylabel('Модель')
    ax.invert_yaxis()
    plt.show()
    print(' ')
    fig, ax = plt.subplots()
    x=data['Цена, руб']
    y=data['Модель']
    ax.plot(x, y)
    ax.set_xlabel('Цена, руб')
    ax.set_ylabel('Модель')
    ax.invert_yaxis()
    plt.show()
    print(' ')
    data.info()
    return data


# Анализ данных
def analysis(data):
    print('\033[1m' + 'Анализ цены' + '\033[0m')
    print(' ')
    # Выделение целевой переменной
    target = data['Цена, руб']

    # Выделение признаков
    features = data.drop(['Модель', 'Страна марки', 'Цена, руб'], axis=1)

    # Разделение данных на тренировочную и тестовую выборки
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.75,
                                                                                random_state=42)

    # Обучение модели линейной регрессии
    model = LinearRegression()
    model.fit(features_train, target_train)
    # Предсказание продаж с помощью обученной модели
    predicted_cost = model.predict(features_test)

    # Построение графика
    fig, ax = plt.subplots()
    ax.plot([0, max(target_test)], [0, max(target_test)], linestyle='--', color='red')  # Пересечение с осью Y = X
    ax.scatter(target_test, predicted_cost)
    ax.set_xlabel('Цена по данным')
    ax.set_ylabel('Предсказанные цены')
    ax.set_title('Сравнение реальных и предсказанных цен')
    ax.set_ylim(bottom=1.5)
    ax.set_xlim(1.5)
    plt.show()
    print(' ')

    # p-values для показателей цены
    p_values = []
    for feature in features_train.columns:
        # сравнение выборки данных(признаки) и выборки для сопоставления(целевой признак)
        t_stat, p_value = stats.ttest_ind(features_train[feature], target_train)
        p_values.append(p_value)
    print("p-values для всех признаков цены:")
    for i, feature in enumerate(features_train.columns):
        print(f"{feature}: {p_values[i]:.15f}")
        print(' ')

    # Calculate R^2
    r2 = r2_score(target_test, predicted_cost)
    print(f"R^2: {r2}")
    # Calculate MAPE
    mape = mean_absolute_percentage_error(target_test, predicted_cost)
    print(f"MAPE: {mape:.2f}%")
    print(' ')
    print(' ')

    print('\033[1m' + 'Анализ Спроса' + '\033[0m')
    print(' ')
    # Выделение целевой переменной (спрос)
    y = data['Продано за 2023']

    # Выделение признаков
    x = data.drop(['Модель', 'Страна марки', 'Продано за 2023'], axis=1)

    # Разделение данных на тренировочную и тестовую выборки
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.75, random_state=84)

    # Обучение модели линейной регрессии
    model = LinearRegression()
    model.fit(x_train, y_train)

    # Предсказание продаж
    predicted_sales = model.predict(x_test)

    # Визуализация результатов
    fig, ax = plt.subplots()
    ax.plot([0, max(y_test)], [0, max(y_test)], linestyle='--', color='red')  # Пересечение с осью Y = X
    ax.scatter(y_test, predicted_sales, color="red")
    ax.set_xlabel('Спрос (Продано за 2023)')
    ax.set_ylabel('Предсказанный спрос')
    ax.set_title('Сравнение реального и предсказанного спроса')
    ax.set_ylim(bottom=0)
    ax.set_xlim(0)
    plt.show()
    print(' ')

    # p-values для показателей спроса
    p_valuec = []
    for featur in x_train.columns:
        # сравнение выборки данных(признаки) и выборки для сопоставления(целевой признак)
        t_stat, p_ = stats.ttest_ind(x_train[featur], y_train)
        p_valuec.append(p_)
    print("p-values для всех признаков спроса:")
    for i, featur in enumerate(x_train.columns):
        print(f"{featur}: {p_valuec[i]:.15f}")
        print(' ')

    # Calculate R^2
    r2 = r2_score(y_test, predicted_sales)
    print(f"R^2: {r2}")
    # Calculate MAPE
    mape = mean_absolute_percentage_error(y_test, predicted_sales)
    print(f"MAPE: {mape:.2f}%")


a = preprocessing(file)
# Анализ данных
analysis(a)

```
